```python
# Import dependency

import pandas as pd
```

```python
# Read csv file

file_location = "DataSource/data_cleaning_challenge.csv"
data_import = pd.read_csv(file_location)
data_import
```

```python
# Dropping unnecessary columns from dataframe. 

drop_extra = data_import.drop(columns = ["Unnamed: 9", "Unnamed: 10"])
drop_extra
```

```python
# Getting rid of Nans.

drop_nans = drop_extra[drop_extra["Row Type"].notnull()]
drop_nans
```

```python
# Iteration to groupings

column_values = []
counter = 0

for i in drop_nans["Row Type"]:
    if "first name" in i:
        counter += 1
    column_values.append(counter)    

len(column_values)
```

```python
# Adding iteration column to dataframe. 

iter_cols = drop_nans
iter_cols["Iteration"] = column_values
iter_cols.head()
```

```python
# Delete row that contains Row Type, Iter Number, Power1, Speed1, Speed2, Electricity, Effort, Weight, Torque. 

drop_extra_column_name_rows = iter_cols[iter_cols["Row Type"] != "Row Type"]
drop_extra_column_name_rows
```

```python
# Selecting out the rows with first name
# Drop the unnecessary columns
# Renaming the columns

name_dataframe = drop_extra_column_name_rows[drop_extra_column_name_rows["Row Type"].str.contains("first name")]
name_dataframe.drop(columns=["Speed1", "Speed2", "Electricity", "Effort", "Weight", "Torque"], inplace = True)
name_dataframe.rename(columns={
    "Row Type": "First Name",
    "Iter Number": "Last Name",
    "Power1": "Date"
}, inplace= True)
name_dataframe["First Name"] = name_dataframe["First Name"].str[12:]
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[11:]
name_dataframe["Date"] = name_dataframe["Date"].str[6:]

name_dataframe

```

```python
# Setting up no name dataframe.

no_name_dataframe = drop_extra_column_name_rows[~drop_extra_column_name_rows["Row Type"].str.contains("first name")]
no_name_dataframe
```

```python
# Merging two dataframe (name_dataframe and no_name_dataframe) into one.

final_output = pd.merge(left = name_dataframe, right = no_name_dataframe, how = "inner", on = "Iteration")
final_output = final_output.drop(columns = ["Iteration","Iter Number", "Row Type"])
final_output
```

```python
# Exporting file to csv
save_to_location = "DataSource/Cleaned_DataSource.csv"
final_output.to_csv(save_to_location, index=False)
```
